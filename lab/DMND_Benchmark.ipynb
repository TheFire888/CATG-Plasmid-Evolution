```python
import polars as pl
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pl.read_csv("../data/DMND_Benchmark.txt", separator=' ', has_header=True)
```

```python
x_vals = df.select(pl.col("time")).to_series().to_numpy()
y_vals = df.select(pl.col("lines")).to_series().to_numpy()
```

```python
m, b = np.polyfit(x_vals, y_vals, 1)
```

```python
plt.plot(x_vals, y_vals, marker='o', linestyle='-')
plt.plot(x_vals, m * x_vals + b, label="Regressão linear")
plt.xlabel("Time (in seconds)")
plt.ylabel("Alignments")
```

```python
print(f"Regressão linear: y = {m:.2e} x + {b:.2e}")
```
